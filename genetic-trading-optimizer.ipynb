```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import warnings
warnings.filterwarnings('ignore')

# Try to import cupy for GPU acceleration, fallback to numpy
try:
    import cupy as cp
    GPU_AVAILABLE = True
    print("GPU acceleration available with CuPy")
except ImportError:
    import numpy as cp
    GPU_AVAILABLE = False
    print("Using CPU (NumPy) - Install CuPy for GPU acceleration")

import random
from typing import List, Tuple, Dict, Any
from dataclasses import dataclass
from enum import Enum
import time

# Set random seeds for reproducibility
np.random.seed(42)
random.seed(42)

class EncodingType(Enum):
    BINARY = "binary"
    REAL_VALUED = "real_valued"
    INTEGER = "integer"
    PERMUTATION = "permutation"
    TREE = "tree"

@dataclass
class GAParameters:
    """Configuration class for Genetic Algorithm parameters"""
    population_size: int = 100
    generations: int = 50
    crossover_rate: float = 0.8
    mutation_rate: float = 0.1
    elitism_rate: float = 0.1
    tournament_size: int = 5
    encoding_type: EncodingType = EncodingType.REAL_VALUED
    use_gpu: bool = False

class TradingStrategy:
    """Trading strategy representation with multiple encoding types"""
    
    def __init__(self, encoding_type: EncodingType = EncodingType.REAL_VALUED):
        self.encoding_type = encoding_type
        self.setup_encoding()
    
    def setup_encoding(self):
        """Setup encoding-specific parameters"""
        if self.encoding_type == EncodingType.BINARY:
            # Binary encoding: 8 bits per parameter
            self.chromosome_length = 8 * 6  # 6 parameters, 8 bits each
        elif self.encoding_type == EncodingType.REAL_VALUED:
            # Real-valued encoding: direct parameter representation
            self.chromosome_length = 6
        elif self.encoding_type == EncodingType.INTEGER:
            # Integer encoding: discrete values
            self.chromosome_length = 6
        elif self.encoding_type == EncodingType.PERMUTATION:
            # Permutation encoding: order of technical indicators
            self.chromosome_length = 5
        else:
            self.chromosome_length = 6
    
    def decode_chromosome(self, chromosome):
        """Decode chromosome to trading parameters"""
        if self.encoding_type == EncodingType.BINARY:
            return self._decode_binary(chromosome)
        elif self.encoding_type == EncodingType.REAL_VALUED:
            return self._decode_real_valued(chromosome)
        elif self.encoding_type == EncodingType.INTEGER:
            return self._decode_integer(chromosome)
        elif self.encoding_type == EncodingType.PERMUTATION:
            return self._decode_permutation(chromosome)
        else:
            return self._decode_real_valued(chromosome)
    
    def _decode_binary(self, chromosome):
        """Decode binary chromosome to parameters"""
        params = {}
        bit_groups = np.split(chromosome, 6)
        
        # Convert binary to decimal and scale to parameter ranges
        params['sma_short'] = int(''.join(map(str, bit_groups[0])), 2) * 50 // 255 + 5  # 5-55
        params['sma_long'] = int(''.join(map(str, bit_groups[1])), 2) * 150 // 255 + 20  # 20-170
        params['rsi_period'] = int(''.join(map(str, bit_groups[2])), 2) * 20 // 255 + 10  # 10-30
        params['rsi_oversold'] = int(''.join(map(str, bit_groups[3])), 2) * 40 // 255 + 20  # 20-60
        params['rsi_overbought'] = int(''.join(map(str, bit_groups[4])), 2) * 30 // 255 + 70  # 70-100
        params['stop_loss'] = int(''.join(map(str, bit_groups[5])), 2) * 0.1 // 255 + 0.01  # 0.01-0.11
        
        return params
    
    def _decode_real_valued(self, chromosome):
        """Decode real-valued chromosome to parameters"""
        params = {}
        # Direct mapping with constraints
        params['sma_short'] = max(5, min(55, int(chromosome[0] * 50 + 5)))
        params['sma_long'] = max(20, min(170, int(chromosome[1] * 150 + 20)))
        params['rsi_period'] = max(10, min(30, int(chromosome[2] * 20 + 10)))
        params['rsi_oversold'] = max(20, min(60, chromosome[3] * 40 + 20))
        params['rsi_overbought'] = max(70, min(100, chromosome[4] * 30 + 70))
        params['stop_loss'] = max(0.01, min(0.11, chromosome[5] * 0.1 + 0.01))
        
        return params
    
    def _decode_integer(self, chromosome):
        """Decode integer chromosome to parameters"""
        params = {}
        params['sma_short'] = max(5, min(55, int(chromosome[0])))
        params['sma_long'] = max(20, min(170, int(chromosome[1])))
        params['rsi_period'] = max(10, min(30, int(chromosome[2])))
        params['rsi_oversold'] = max(20, min(60, chromosome[3]))
        params['rsi_overbought'] = max(70, min(100, chromosome[4]))
        params['stop_loss'] = max(0.01, min(0.11, chromosome[5]))
        
        return params
    
    def _decode_permutation(self, chromosome):
        """Decode permutation chromosome (order of indicators)"""
        # This is a simplified version - in practice, you'd use the permutation
        # to determine the order/priority of technical indicators
        indicator_order = np.argsort(chromosome)
        params = {
            'sma_short': 10 + indicator_order[0] * 5,
            'sma_long': 50 + indicator_order[1] * 20,
            'rsi_period': 14 + indicator_order[2] * 2,
            'rsi_oversold': 30,
            'rsi_overbought': 70,
            'stop_loss': 0.05
        }
        return params

class GeneticAlgorithm:
    """Main Genetic Algorithm implementation with multiple encoding support"""
    
    def __init__(self, params: GAParameters):
        self.params = params
        self.strategy = TradingStrategy(params.encoding_type)
        self.population = []
        self.fitness_history = []
        self.best_fitness_history = []
        self.diversity_history = []
        self.generation = 0
        
        # Initialize array backend based on GPU availability
        if params.use_gpu and GPU_AVAILABLE:
            self.xp = cp
            print("Using GPU acceleration")
        else:
            self.xp = np
            print("Using CPU computation")
    
    def initialize_population(self):
        """Initialize population based on encoding type"""
        self.population = []
        
        for _ in range(self.params.population_size):
            if self.params.encoding_type == EncodingType.BINARY:
                chromosome = self.xp.random.randint(0, 2, self.strategy.chromosome_length)
            elif self.params.encoding_type == EncodingType.REAL_VALUED:
                chromosome = self.xp.random.random(self.strategy.chromosome_length)
            elif self.params.encoding_type == EncodingType.INTEGER:
                chromosome = self.xp.random.randint(1, 100, self.strategy.chromosome_length)
            elif self.params.encoding_type == EncodingType.PERMUTATION:
                chromosome = self.xp.random.permutation(self.strategy.chromosome_length)
            else:
                chromosome = self.xp.random.random(self.strategy.chromosome_length)
            
            self.population.append(chromosome)
    
    def evaluate_fitness(self, chromosome, price_data):
        """Evaluate fitness of a chromosome using trading strategy"""
        # Convert to numpy if using GPU
        if self.params.use_gpu and GPU_AVAILABLE:
            chromosome = cp.asnumpy(chromosome)
        
        params = self.strategy.decode_chromosome(chromosome)
        
        # Calculate technical indicators
        sma_short = self.calculate_sma(price_data, params['sma_short'])
        sma_long = self.calculate_sma(price_data, params['sma_long'])
        rsi = self.calculate_rsi(price_data, params['rsi_period'])
        
        # Generate trading signals
        signals = self.generate_signals(price_data, sma_short, sma_long, rsi, params)
        
        # Calculate portfolio performance
        returns = self.calculate_returns(price_data, signals, params['stop_loss'])
        
        # Fitness function: Sharpe ratio + total return - max drawdown
        if len(returns) > 0:
            total_return = np.sum(returns)
            volatility = np.std(returns) if np.std(returns) > 0 else 0.001
            sharpe_ratio = total_return / volatility
            max_drawdown = self.calculate_max_drawdown(returns)
            
            fitness = sharpe_ratio + total_return * 0.1 - max_drawdown * 0.5
        else:
            fitness = -1000  # Penalty for no trades
        
        return fitness
    
    def calculate_sma(self, prices, period):
        """Calculate Simple Moving Average"""
        return pd.Series(prices).rolling(window=period).mean().fillna(0)
    
    def calculate_rsi(self, prices, period):
        """Calculate Relative Strength Index"""
        delta = pd.Series(prices).diff()
        gain = (delta.where(delta > 0, 0)).rolling(window=period).mean()
        loss = (-delta.where(delta < 0, 0)).rolling(window=period).mean()
        rs = gain / loss
        rsi = 100 - (100 / (1 + rs))
        return rsi.fillna(50)
    
    def generate_signals(self, prices, sma_short, sma_long, rsi, params):
        """Generate buy/sell signals based on technical indicators"""
        signals = []
        position = 0  # 0: no position, 1: long, -1: short
        
        for i in range(len(prices)):
            signal = 0
            
            # SMA crossover strategy
            if i > 0:
                if sma_short.iloc[i] > sma_long.iloc[i] and sma_short.iloc[i-1] <= sma_long.iloc[i-1]:
                    signal = 1  # Buy signal
                elif sma_short.iloc[i] < sma_long.iloc[i] and sma_short.iloc[i-1] >= sma_long.iloc[i-1]:
                    signal = -1  # Sell signal
            
            # RSI confirmation
            if signal == 1 and rsi.iloc[i] < params['rsi_overbought']:
                signal = 1
            elif signal == -1 and rsi.iloc[i] > params['rsi_oversold']:
                signal = -1
            else:
                signal = 0
            
            signals.append(signal)
        
        return signals
    
    def calculate_returns(self, prices, signals, stop_loss):
        """Calculate portfolio returns based on signals"""
        returns = []
        position = 0
        entry_price = 0
        
        for i in range(1, len(prices)):
            daily_return = 0
            
            if signals[i] == 1 and position == 0:  # Buy
                position = 1
                entry_price = prices[i]
            elif signals[i] == -1 and position == 1:  # Sell
                daily_return = (prices[i] - entry_price) / entry_price
                position = 0
                entry_price = 0
            elif position == 1:  # Holding long position
                current_return = (prices[i] - entry_price) / entry_price
                if current_return <= -stop_loss:  # Stop loss triggered
                    daily_return = -stop_loss
                    position = 0
                    entry_price = 0
            
            returns.append(daily_return)
        
        return returns
    
    def calculate_max_drawdown(self, returns):
        """Calculate maximum drawdown"""
        cumulative = np.cumprod(1 + np.array(returns))
        running_max = np.maximum.accumulate(cumulative)
        drawdown = (cumulative - running_max) / running_max
        return abs(np.min(drawdown))
    
    def selection(self, fitness_scores):
        """Tournament selection"""
        selected = []
        for _ in range(self.params.population_size):
            tournament = np.random.choice(len(fitness_scores), self.params.tournament_size, replace=False)
            winner = tournament[np.argmax([fitness_scores[i] for i in tournament])]
            selected.append(self.population[winner].copy())
        return selected
    
    def crossover(self, parent1, parent2):
        """Crossover operation based on encoding type"""
        if self.params.encoding_type == EncodingType.BINARY:
            return self._binary_crossover(parent1, parent2)
        elif self.params.encoding_type == EncodingType.REAL_VALUED:
            return self._real_crossover(parent1, parent2)
        elif self.params.encoding_type == EncodingType.INTEGER:
            return self._integer_crossover(parent1, parent2)
        elif self.params.encoding_type == EncodingType.PERMUTATION:
            return self._permutation_crossover(parent1, parent2)
        else:
            return self._real_crossover(parent1, parent2)
    
    def _binary_crossover(self, parent1, parent2):
        """Single-point crossover for binary encoding"""
        point = np.random.randint(1, len(parent1))
        child1 = np.concatenate([parent1[:point], parent2[point:]])
        child2 = np.concatenate([parent2[:point], parent1[point:]])
        return child1, child2
    
    def _real_crossover(self, parent1, parent2):
        """Blend crossover for real-valued encoding"""
        alpha = 0.5
        child1 = np.zeros_like(parent1)
        child2 = np.zeros_like(parent2)
        
        for i in range(len(parent1)):
            if np.random.random() < 0.5:
                child1[i] = parent1[i]
                child2[i] = parent2[i]
            else:
                child1[i] = parent2[i]
                child2[i] = parent1[i]
        
        return child1, child2
    
    def _integer_crossover(self, parent1, parent2):
        """Uniform crossover for integer encoding"""
        child1 = np.zeros_like(parent1)
        child2 = np.zeros_like(parent2)
        
        for i in range(len(parent1)):
            if np.random.random() < 0.5:
                child1[i] = parent1[i]
                child2[i] = parent2[i]
            else:
                child1[i] = parent2[i]
                child2[i] = parent1[i]
        
        return child1, child2
    
    def _permutation_crossover(self, parent1, parent2):
        """Order crossover for permutation encoding"""
        size = len(parent1)
        start, end = sorted(np.random.choice(size, 2, replace=False))
        
        child1 = np.full(size, -1)
        child2 = np.full(size, -1)
        
        # Copy segment from parent1 to child1
        child1[start:end] = parent1[start:end]
        child2[start:end] = parent2[start:end]
        
        # Fill remaining positions
        self._fill_permutation_child(child1, parent2, start, end)
        self._fill_permutation_child(child2, parent1, start, end)
        
        return child1, child2
    
    def _fill_permutation_child(self, child, parent, start, end):
        """Helper function for permutation crossover"""
        remaining = [x for x in parent if x not in child[start:end]]
        idx = 0
        for i in range(len(child)):
            if child[i] == -1:
                child[i] = remaining[idx]
                idx += 1
    
    def mutate(self, chromosome):
        """Mutation operation based on encoding type"""
        if self.params.encoding_type == EncodingType.BINARY:
            return self._binary_mutation(chromosome)
        elif self.params.encoding_type == EncodingType.REAL_VALUED:
            return self._real_mutation(chromosome)
        elif self.params.encoding_type == EncodingType.INTEGER:
            return self._integer_mutation(chromosome)
        elif self.params.encoding_type == EncodingType.PERMUTATION:
            return self._permutation_mutation(chromosome)
        else:
            return self._real_mutation(chromosome)
    
    def _binary_mutation(self, chromosome):
        """Bit flip mutation for binary encoding"""
        mutated = chromosome.copy()
        for i in range(len(mutated)):
            if np.random.random() < self.params.mutation_rate:
                mutated[i] = 1 - mutated[i]
        return mutated
    
    def _real_mutation(self, chromosome):
        """Gaussian mutation for real-valued encoding"""
        mutated = chromosome.copy()
        for i in range(len(mutated)):
            if np.random.random() < self.params.mutation_rate:
                mutated[i] += np.random.normal(0, 0.1)
                mutated[i] = np.clip(mutated[i], 0, 1)
        return mutated
    
    def _integer_mutation(self, chromosome):
        """Random reset mutation for integer encoding"""
        mutated = chromosome.copy()
        for i in range(len(mutated)):
            if np.random.random() < self.params.mutation_rate:
                mutated[i] = np.random.randint(1, 100)
        return mutated
    
    def _permutation_mutation(self, chromosome):
        """Swap mutation for permutation encoding"""
        mutated = chromosome.copy()
        if np.random.random() < self.params.mutation_rate:
            i, j = np.random.choice(len(mutated), 2, replace=False)
            mutated[i], mutated[j] = mutated[j], mutated[i]
        return mutated
    
    def calculate_diversity(self):
        """Calculate population diversity"""
        if len(self.population) < 2:
            return 0
        
        total_distance = 0
        count = 0
        
        for i in range(len(self.population)):
            for j in range(i+1, len(self.population)):
                # Convert to numpy if using GPU
                ind1 = cp.asnumpy(self.population[i]) if self.params.use_gpu and GPU_AVAILABLE else self.population[i]
                ind2 = cp.asnumpy(self.population[j]) if self.params.use_gpu and GPU_AVAILABLE else self.population[j]
                
                distance = np.linalg.norm(ind1 - ind2)
                total_distance += distance
                count += 1
        
        return total_distance / count if count > 0 else 0
    
    def evolve(self, price_data, verbose=True):
        """Main evolution loop"""
        # Initialize population
        self.initialize_population()
        
        for generation in range(self.params.generations):
            self.generation = generation
            
            # Evaluate fitness
            fitness_scores = []
            for chromosome in self.population:
                fitness = self.evaluate_fitness(chromosome, price_data)
                fitness_scores.append(fitness)
            
            # Track statistics
            self.fitness_history.append(np.mean(fitness_scores))
            self.best_fitness_history.append(np.max(fitness_scores))
            self.diversity_history.append(self.calculate_diversity())
            
            if verbose and generation % 10 == 0:
                print(f"Generation {generation}: Best Fitness = {np.max(fitness_scores):.4f}, "
                      f"Avg Fitness = {np.mean(fitness_scores):.4f}, "
                      f"Diversity = {self.diversity_history[-1]:.4f}")
            
            # Selection
            selected = self.selection(fitness_scores)
            
            # Elitism: Keep best individuals
            elite_count = int(self.params.population_size * self.params.elitism_rate)
            elite_indices = np.argsort(fitness_scores)[-elite_count:]
            elites = [self.population[i].copy() for i in elite_indices]
            
            # Crossover and mutation
            new_population = elites.copy()
            
            while len(new_population) < self.params.population_size:
                parent1, parent2 = np.random.choice(len(selected), 2, replace=False)
                
                if np.random.random() < self.params.crossover_rate:
                    child1, child2 = self.crossover(selected[parent1], selected[parent2])
                else:
                    child1, child2 = selected[parent1].copy(), selected[parent2].copy()
                
                child1 = self.mutate(child1)
                child2 = self.mutate(child2)
                
                new_population.extend([child1, child2])
            
            # Trim population to exact size
            self.population = new_population[:self.params.population_size]
        
        # Return best solution
        final_fitness = [self.evaluate_fitness(chromosome, price_data) for chromosome in self.population]
        best_idx = np.argmax(final_fitness)
        best_chromosome = self.population[best_idx]
        
        # Convert to numpy if using GPU
        if self.params.use_gpu and GPU_AVAILABLE:
            best_chromosome = cp.asnumpy(best_chromosome)
        
        return best_chromosome, np.max(final_fitness)

# Generate sample stock price data
def generate_sample_data(days=500):
    """Generate synthetic stock price data"""
    np.random.seed(42)
    
    # Generate price data with trend and volatility
    base_price = 100
    trend = 0.0005  # Daily trend
    volatility = 0.02  # Daily volatility
    
    prices = [base_price]
    for i in range(days):
        change = np.random.normal(trend, volatility)
        new_price = prices[-1] * (1 + change)
        prices.append(new_price)
    
    dates = pd.date_range('2022-01-01', periods=days+1, freq='D')
    
    return pd.DataFrame({
        'date': dates,
        'price': prices
    })

# Interactive parameter adjustment class
class InteractiveGA:
    """Interactive Genetic Algorithm with parameter adjustment"""
    
    def __init__(self):
        self.data = generate_sample_data()
        self.current_params = GAParameters()
        self.results = {}
        self.ga = None
    
    def run_experiment(self, **kwargs):
        """Run GA experiment with given parameters"""
        # Update parameters
        for key, value in kwargs.items():
            if hasattr(self.current_params, key):
                setattr(self.current_params, key, value)
        
        print(f"Running experiment with parameters:")
        print(f"Population Size: {self.current_params.population_size}")
        print(f"Generations: {self.current_params.generations}")
        print(f"Crossover Rate: {self.current_params.crossover_rate}")
        print(f"Mutation Rate: {self.current_params.mutation_rate}")
        print(f"Encoding Type: {self.current_params.encoding_type.value}")
        print(f"Use GPU: {self.current_params.use_gpu}")
        print("-" * 50)
        
        # Run genetic algorithm
        start_time = time.time()
        self.ga = GeneticAlgorithm(self.current_params)
        best_solution, best_fitness = self.ga.evolve(self.data['price'].values)
        end_time = time.time()
        
        # Store results
        experiment_key = f"{self.current_params.encoding_type.value}_{self.current_params.population_size}_{self.current_params.generations}"
        self.results[experiment_key] = {
            'best_solution': best_solution,
            'best_fitness': best_fitness,
            'fitness_history': self.ga.fitness_history,
            'best_fitness_history': self.ga.best_fitness_history,
            'diversity_history': self.ga.diversity_history,
            'execution_time': end_time - start_time,
            'parameters': self.current_params
        }
        
        print(f"Experiment completed in {end_time - start_time:.2f} seconds")
        print(f"Best fitness: {best_fitness:.4f}")
        
        # Decode and display best solution
        best_params = self.ga.strategy.decode_chromosome(best_solution)
        print(f"Best trading parameters: {best_params}")
        
        return experiment_key
    
    def visualize_results(self, experiment_keys=None):
        """Visualize results from experiments"""
        if experiment_keys is None:
            experiment_keys = list(self.results.keys())
        
        if not experiment_keys:
            print("No experiments to visualize")
            return
        
        # Create comprehensive visualization
        fig, axes = plt.subplots(2, 3, figsize=(18, 12))
        fig.suptitle('Genetic Algorithm Optimization Results', fontsize=16)
        
        # 1. Fitness Evolution
        ax1 = axes[0, 0]
        for key in experiment_keys:
            result = self.results[key]
            ax1.plot(result['fitness_history'], label=f"{key} (Avg)", alpha=0.7)
            ax1.plot(result['best_fitness_history'], label=f"{key} (Best)", linestyle='--', alpha=0.7)
        ax1.set_xlabel('Generation')
        ax1.set_ylabel('Fitness')
        ax1.set_title('Fitness Evolution')
        ax1.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        ax1.grid(True, alpha=0.3)
        
        # 2. Diversity Evolution
        ax2 = axes[0, 1]
        for key in experiment_keys:
            result = self.results[key]
            ax2.plot(result['diversity_history'], label=key, alpha=0.7)
        ax2.set_xlabel('Generation')
        ax2.set_ylabel('Population Diversity')
        ax2.set_title('Population Diversity Evolution')
        ax2.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        ax2.grid(True, alpha=0.3)
        
        # 3. Final Performance Comparison
        ax3 = axes[0, 2]
        final_fitness = [self.results[key]['best_fitness'] for key in experiment_keys]
        execution_times = [self.results[key]['execution_time'] for key in experiment_keys]
        
        scatter = ax3.scatter(execution_times, final_fitness, 
                            s=100, alpha=0.7, c=range(len(experiment_keys)), cmap='viridis')
        ax3.set_xlabel('Execution Time (s)')
        ax3.set_ylabel('Best Fitness')
        ax3.set_title('Performance vs Execution Time')
        ax3.grid(True, alpha=0.3)
        
        # Add labels for each point
        for i, key in enumerate(experiment_keys):
            ax3.annotate(key.split('_')[0], 
                        (execution_times[i], final_fitness[i]),
                        xytext=(5, 5), textcoords='offset points', fontsize=8)
        
        # 4. Parameter Sensitivity Analysis
        ax4 = axes[1, 0]
        if len(experiment_keys) > 1:
            # Group by encoding type
            encoding_performance = {}
            for key in experiment_keys:
                encoding = key.split('_')[0]
                if encoding not in encoding_performance:
                    encoding_performance[encoding] = []
                encoding_performance[encoding].append(self.results[key]['best_fitness'])
            
            encodings = list(encoding_performance.keys())
            performances = [np.mean(encoding_performance[enc]) for enc in encodings]
            
            ax4.bar(encodings, performances, alpha=0.7)
            ax4.set_xlabel('Encoding Type')
            ax4.set_ylabel('Average Best Fitness')
            ax4.set_title('Performance by Encoding Type')
            plt.setp(ax4.get_xticklabels(), rotation=45, ha='right')
        
        # 5. Trading Strategy Visualization
        ax5 = axes[1, 1]
        if experiment_keys:
            # Use the best experiment
            best_key = max(experiment_keys, key=lambda k: self.results[k]['best_fitness'])
            best_result = self.results[best_key]
            
            # Simulate trading with best strategy
            best_params = self.ga.strategy.decode_chromosome(best_result['best_solution'])
            
            # Plot price and signals
            prices = self.data['price'].values[-100:]  # Last 100 days
            ax5.plot(prices, label='Price', alpha=0.7)
            
            # Calculate and plot moving averages
            sma_short = pd.Series(prices).rolling(window=min(best_params['sma_short'], len(prices))).mean()
            sma_long = pd.Series(prices).rolling(window=min(best_params['sma_long'], len(prices))).mean()
            
            ax5.plot(sma_short, label=f'SMA{best_params["sma_short"]}', alpha=0.7)
            ax5.plot(sma_long, label=f'SMA{best_params["sma_long"]}', alpha=0.7)
            
            ax5.set_xlabel('Days')
            ax5.set_ylabel('Price')
            ax5.set_title(f'Best Trading Strategy ({best_key})')
            ax5.legend()
            ax5.grid(True, alpha=0.3)
        
        # 6. Convergence Analysis
        ax6 = axes[1, 2]
        for key in experiment_keys:
            result = self.results[key]
            # Calculate convergence rate (how quickly fitness improves)
            fitness_diff = np.diff(result['best_fitness_history'])
            convergence_rate = np.cumsum(np.abs(fitness_diff))
            ax6.plot(convergence_rate, label=key, alpha=0.7)
        
        ax6.set_xlabel('Generation')
        ax6.set_ylabel('Cumulative Fitness Change')
        ax6.set_title('Convergence Rate Analysis')
        ax6.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        ax6.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()
        
        # Print summary statistics
        print("\n" + "="*80)
        print("EXPERIMENT SUMMARY")
        print("="*80)
        
        for key in experiment_keys:
            result = self.results[key]
            print(f"\nExperiment: {key}")
            print(f"  Best Fitness: {result['best_fitness']:.4f}")
            print(f"  Execution Time: {result['execution_time']:.2f}s")
            print(f"  Final Diversity: {result['diversity_history'][-1]:.4f}")
            print(f"  Convergence: {result['best_fitness_history'][-1] - result['best_fitness_history'][0]:.4f}")
    
    def compare_encodings(self):
        """Compare different encoding types"""
        print("Running comparison of different encoding types...")
        
        encoding_types = [
            EncodingType.BINARY,
            EncodingType.REAL_VALUED,
            EncodingType.INTEGER,
            EncodingType.PERMUTATION
        ]
        
        experiment_keys = []
        
        for encoding in encoding_types:
            print(f"\nTesting {encoding.value} encoding...")
            key = self.run_experiment(
                encoding_type=encoding,
                population_size=50,
                generations=30,
                crossover_rate=0.8,
                mutation_rate=0.1
            )
            experiment_keys.append(key)
        
        # Visualize comparison
        self.visualize_results(experiment_keys)
        
        return experiment_keys
        
    def parameter_sensitivity_analysis(self):
        """Analyze sensitivity to different parameters"""
        print("Running parameter sensitivity analysis...")
        
        base_params = {
            'encoding_type': EncodingType.REAL_VALUED,
            'generations': 25,
            'crossover_rate': 0.8,
            'mutation_rate': 0.1
        }
        
        experiment_keys = []
        
        # Test different population sizes
        for pop_size in [20, 50, 100, 200]:
            print(f"\nTesting population size: {pop_size}")
            key = self.run_experiment(population_size=pop_size, **base_params)
            experiment_keys.append(key)
        
        # Test different mutation rates - Remove mutation_rate from base_params
        base_params_no_mutation = {k: v for k, v in base_params.items() if k != 'mutation_rate'}
        for mut_rate in [0.05, 0.1, 0.2, 0.3]:
            print(f"\nTesting mutation rate: {mut_rate}")
            key = self.run_experiment(mutation_rate=mut_rate, population_size=50, **base_params_no_mutation)
            experiment_keys.append(key)
        
        # Test different crossover rates - Remove crossover_rate from base_params
        base_params_no_crossover = {k: v for k, v in base_params.items() if k != 'crossover_rate'}
        for cross_rate in [0.6, 0.7, 0.8, 0.9]:
            print(f"\nTesting crossover rate: {cross_rate}")
            key = self.run_experiment(crossover_rate=cross_rate, population_size=50, **base_params_no_crossover)
            experiment_keys.append(key)
        
        self.visualize_results(experiment_keys)
        return experiment_keys
    
    def gpu_vs_cpu_comparison(self):
        """Compare GPU vs CPU performance"""
        if not GPU_AVAILABLE:
            print("GPU not available for comparison")
            return
        
        print("Comparing GPU vs CPU performance...")
        
        base_params = {
            'encoding_type': EncodingType.REAL_VALUED,
            'population_size': 100,
            'generations': 50,
            'crossover_rate': 0.8,
            'mutation_rate': 0.1
        }
        
        experiment_keys = []
        
        # CPU run
        print("\nRunning on CPU...")
        key_cpu = self.run_experiment(use_gpu=False, **base_params)
        experiment_keys.append(key_cpu)
        
        # GPU run
        print("\nRunning on GPU...")
        key_gpu = self.run_experiment(use_gpu=True, **base_params)
        experiment_keys.append(key_gpu)
        
        # Compare results
        cpu_time = self.results[key_cpu]['execution_time']
        gpu_time = self.results[key_gpu]['execution_time']
        speedup = cpu_time / gpu_time
        
        print(f"\nPerformance Comparison:")
        print(f"CPU Time: {cpu_time:.2f}s")
        print(f"GPU Time: {gpu_time:.2f}s")
        print(f"Speedup: {speedup:.2f}x")
        
        self.visualize_results(experiment_keys)
        return experiment_keys

# Advanced visualization functions
def create_animated_evolution(ga_instance, save_path=None):
    """Create animated visualization of GA evolution"""
    if not hasattr(ga_instance, 'fitness_history'):
        print("No evolution history available")
        return
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    def animate(frame):
        ax1.clear()
        ax2.clear()
        
        # Plot fitness evolution up to current frame
        generations = range(frame + 1)
        avg_fitness = ga_instance.fitness_history[:frame + 1]
        best_fitness = ga_instance.best_fitness_history[:frame + 1]
        
        ax1.plot(generations, avg_fitness, 'b-', label='Average Fitness', linewidth=2)
        ax1.plot(generations, best_fitness, 'r-', label='Best Fitness', linewidth=2)
        ax1.set_xlabel('Generation')
        ax1.set_ylabel('Fitness')
        ax1.set_title(f'Fitness Evolution (Generation {frame})')
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        # Plot diversity evolution
        diversity = ga_instance.diversity_history[:frame + 1]
        ax2.plot(generations, diversity, 'g-', label='Population Diversity', linewidth=2)
        ax2.set_xlabel('Generation')
        ax2.set_ylabel('Diversity')
        ax2.set_title(f'Population Diversity (Generation {frame})')
        ax2.legend()
        ax2.grid(True, alpha=0.3)
        
        # Add statistics text
        if frame < len(avg_fitness):
            stats_text = f"Gen {frame}: Avg={avg_fitness[frame]:.3f}, Best={best_fitness[frame]:.3f}, Div={diversity[frame]:.3f}"
            fig.suptitle(stats_text, fontsize=12)
    
    anim = FuncAnimation(fig, animate, frames=len(ga_instance.fitness_history), 
                        interval=200, repeat=True, blit=False)
    
    if save_path:
        anim.save(save_path, writer='pillow', fps=5)
        print(f"Animation saved to {save_path}")
    
    plt.tight_layout()
    plt.show()
    
    return anim

def plot_parameter_landscape(interactive_ga, parameter_name, values, fixed_params=None):
    """Plot fitness landscape for a specific parameter"""
    if fixed_params is None:
        fixed_params = {}
    
    results = []
    
    print(f"Analyzing {parameter_name} landscape...")
    
    for value in values:
        params = fixed_params.copy()
        params[parameter_name] = value
        
        print(f"Testing {parameter_name} = {value}")
        key = interactive_ga.run_experiment(**params)
        results.append(interactive_ga.results[key]['best_fitness'])
    
    # Create visualization
    plt.figure(figsize=(12, 8))
    
    # Main plot
    plt.subplot(2, 2, 1)
    plt.plot(values, results, 'bo-', linewidth=2, markersize=8)
    plt.xlabel(parameter_name.replace('_', ' ').title())
    plt.ylabel('Best Fitness')
    plt.title(f'Fitness Landscape: {parameter_name}')
    plt.grid(True, alpha=0.3)
    
    # Find optimal value
    optimal_idx = np.argmax(results)
    optimal_value = values[optimal_idx]
    optimal_fitness = results[optimal_idx]
    
    plt.axvline(x=optimal_value, color='r', linestyle='--', alpha=0.7, 
                label=f'Optimal: {optimal_value}')
    plt.legend()
    
    # Distribution plot
    plt.subplot(2, 2, 2)
    plt.hist(results, bins=min(10, len(results)), alpha=0.7, edgecolor='black')
    plt.xlabel('Fitness')
    plt.ylabel('Frequency')
    plt.title('Fitness Distribution')
    plt.axvline(x=optimal_fitness, color='r', linestyle='--', alpha=0.7)
    
    # Sensitivity analysis
    plt.subplot(2, 2, 3)
    sensitivity = np.abs(np.diff(results))
    sensitivity_x = [(values[i] + values[i+1]) / 2 for i in range(len(values)-1)]
    plt.plot(sensitivity_x, sensitivity, 'go-', linewidth=2, markersize=6)
    plt.xlabel(parameter_name.replace('_', ' ').title())
    plt.ylabel('Fitness Sensitivity')
    plt.title('Parameter Sensitivity')
    plt.grid(True, alpha=0.3)
    
    # Summary statistics
    plt.subplot(2, 2, 4)
    stats_text = f"""
    Parameter: {parameter_name}
    Optimal Value: {optimal_value}
    Optimal Fitness: {optimal_fitness:.4f}
    Mean Fitness: {np.mean(results):.4f}
    Std Fitness: {np.std(results):.4f}
    Range: {np.max(results) - np.min(results):.4f}
    """
    plt.text(0.1, 0.5, stats_text, transform=plt.gca().transAxes, 
             fontsize=10, verticalalignment='center',
             bbox=dict(boxstyle="round,pad=0.3", facecolor="lightblue", alpha=0.7))
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()
    
    return optimal_value, optimal_fitness

# Main execution and demonstration
def main():
    """Main demonstration of the Genetic Algorithm"""
    print("ðŸ§¬ COMPREHENSIVE GENETIC ALGORITHM DEMONSTRATION ðŸ§¬")
    print("="*60)
    
    # Create interactive GA instance
    interactive_ga = InteractiveGA()
    
    # 1. Basic demonstration
    print("\n1. BASIC GENETIC ALGORITHM RUN")
    print("-" * 40)
    basic_key = interactive_ga.run_experiment(
        population_size=100,
        generations=50,
        crossover_rate=0.8,
        mutation_rate=0.1,
        encoding_type=EncodingType.REAL_VALUED,
        use_gpu=False
    )
    
    # 2. Compare different encoding types
    print("\n2. ENCODING TYPE COMPARISON")
    print("-" * 40)
    encoding_keys = interactive_ga.compare_encodings()
    
    # 3. Parameter sensitivity analysis
    print("\n3. PARAMETER SENSITIVITY ANALYSIS")
    print("-" * 40)
    sensitivity_keys = interactive_ga.parameter_sensitivity_analysis()
    
    # 4. GPU vs CPU comparison (if available)
    print("\n4. GPU vs CPU COMPARISON")
    print("-" * 40)
    if GPU_AVAILABLE:
        gpu_keys = interactive_ga.gpu_vs_cpu_comparison()
    else:
        print("GPU not available - skipping comparison")
    
    # 5. Advanced parameter landscape analysis
    print("\n5. ADVANCED PARAMETER LANDSCAPE")
    print("-" * 40)
    
    # Analyze population size landscape
    pop_sizes = [20, 40, 60, 80, 100, 120, 150, 200]
    optimal_pop_size, optimal_fitness = plot_parameter_landscape(
        interactive_ga, 
        'population_size', 
        pop_sizes,
        {'encoding_type': EncodingType.REAL_VALUED, 'generations': 30}
    )
    
    # Analyze mutation rate landscape
    mutation_rates = [0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4]
    optimal_mutation_rate, _ = plot_parameter_landscape(
        interactive_ga,
        'mutation_rate',
        mutation_rates,
        {'encoding_type': EncodingType.REAL_VALUED, 'generations': 30, 'population_size': 100}
    )
    
    # 6. Create animated evolution
    print("\n6. ANIMATED EVOLUTION")
    print("-" * 40)
    
    # Run a new experiment for animation
    animation_key = interactive_ga.run_experiment(
        population_size=100,
        generations=50,
        encoding_type=EncodingType.REAL_VALUED,
        crossover_rate=0.8,
        mutation_rate=0.1
    )
    
    # Create animation
    anim = create_animated_evolution(interactive_ga.ga)
    
    # 7. Final optimized run
    print("\n7. FINAL OPTIMIZED RUN")
    print("-" * 40)
    
    optimal_key = interactive_ga.run_experiment(
        population_size=int(optimal_pop_size),
        generations=100,
        mutation_rate=optimal_mutation_rate,
        crossover_rate=0.8,
        encoding_type=EncodingType.REAL_VALUED,
        use_gpu=GPU_AVAILABLE
    )
    
    # 8. Comprehensive results visualization
    print("\n8. COMPREHENSIVE RESULTS")
    print("-" * 40)
    
    all_keys = [basic_key] + encoding_keys + sensitivity_keys
    if GPU_AVAILABLE:
        all_keys += gpu_keys
    all_keys.append(optimal_key)
    
    interactive_ga.visualize_results(all_keys)
    
    # 9. Final summary and recommendations
    print("\n9. FINAL SUMMARY AND RECOMMENDATIONS")
    print("-" * 40)
    
    best_result = max(interactive_ga.results.values(), key=lambda x: x['best_fitness'])
    # best_key = [k for k, v in interactive_ga.results.items() if v == best_result][0]
    best_key = [k for k, v in interactive_ga.results.items() if v is best_result][0]
    
    print(f"\nðŸ† BEST PERFORMING CONFIGURATION:")
    print(f"Experiment: {best_key}")
    print(f"Best Fitness: {best_result['best_fitness']:.4f}")
    print(f"Execution Time: {best_result['execution_time']:.2f}s")
    
    # Decode best solution
    best_params = interactive_ga.ga.strategy.decode_chromosome(best_result['best_solution'])
    print(f"\nðŸ“Š OPTIMAL TRADING STRATEGY PARAMETERS:")
    for param, value in best_params.items():
        print(f"  {param}: {value}")
    
    print(f"\nðŸ’¡ RECOMMENDATIONS:")
    print(f"  â€¢ Optimal Population Size: {optimal_pop_size}")
    print(f"  â€¢ Optimal Mutation Rate: {optimal_mutation_rate}")
    print(f"  â€¢ Best Encoding: {best_key.split('_')[0]}")
    print(f"  â€¢ Use GPU: {'Yes' if GPU_AVAILABLE else 'No (install CuPy)'}")
    
    print(f"\nðŸŽ¯ PERFORMANCE METRICS:")
    print(f"  â€¢ Convergence Rate: {(best_result['best_fitness_history'][-1] - best_result['best_fitness_history'][0]):.4f}")
    print(f"  â€¢ Final Diversity: {best_result['diversity_history'][-1]:.4f}")
    print(f"  â€¢ Generations to Convergence: {len(best_result['best_fitness_history'])}")
    
    return interactive_ga

# Interactive parameter adjustment functions
def adjust_parameters_interactively():
    """Interactive parameter adjustment interface"""
    print("\nðŸ”§ INTERACTIVE PARAMETER ADJUSTMENT")
    print("="*50)
    print("Adjust parameters and see immediate results!")
    
    interactive_ga = InteractiveGA()
    
    # Parameter options
    encoding_options = {
        1: EncodingType.BINARY,
        2: EncodingType.REAL_VALUED,
        3: EncodingType.INTEGER,
        4: EncodingType.PERMUTATION
    }
    
    while True:
        print("\n" + "="*50)
        print("CURRENT PARAMETERS:")
        print(f"1. Population Size: {interactive_ga.current_params.population_size}")
        print(f"2. Generations: {interactive_ga.current_params.generations}")
        print(f"3. Crossover Rate: {interactive_ga.current_params.crossover_rate}")
        print(f"4. Mutation Rate: {interactive_ga.current_params.mutation_rate}")
        print(f"5. Encoding Type: {interactive_ga.current_params.encoding_type.value}")
        print(f"6. Use GPU: {interactive_ga.current_params.use_gpu}")
        
        print("\nOPTIONS:")
        print("1-6: Change parameter")
        print("R: Run experiment")
        print("V: View results")
        print("C: Compare all runs")
        print("Q: Quit")
        
        choice = input("\nEnter your choice: ").upper()
        
        if choice == 'Q':
            break
        elif choice == 'R':
            key = interactive_ga.run_experiment()
            print(f"Experiment completed: {key}")
        elif choice == 'V':
            if interactive_ga.results:
                interactive_ga.visualize_results()
            else:
                print("No results to display. Run an experiment first.")
        elif choice == 'C':
            if len(interactive_ga.results) > 1:
                interactive_ga.visualize_results(list(interactive_ga.results.keys()))
            else:
                print("Need at least 2 experiments to compare.")
        elif choice in ['1', '2', '3', '4', '5', '6']:
            param_idx = int(choice)
            
            if param_idx == 1:
                new_value = int(input("Enter new population size (10-500): "))
                interactive_ga.current_params.population_size = max(10, min(500, new_value))
            elif param_idx == 2:
                new_value = int(input("Enter new generations (10-200): "))
                interactive_ga.current_params.generations = max(10, min(200, new_value))
            elif param_idx == 3:
                new_value = float(input("Enter new crossover rate (0.0-1.0): "))
                interactive_ga.current_params.crossover_rate = max(0.0, min(1.0, new_value))
            elif param_idx == 4:
                new_value = float(input("Enter new mutation rate (0.0-1.0): "))
                interactive_ga.current_params.mutation_rate = max(0.0, min(1.0, new_value))
            elif param_idx == 5:
                print("Encoding types:")
                for i, enc in encoding_options.items():
                    print(f"{i}: {enc.value}")
                enc_choice = int(input("Select encoding type (1-4): "))
                if enc_choice in encoding_options:
                    interactive_ga.current_params.encoding_type = encoding_options[enc_choice]
            elif param_idx == 6:
                new_value = input("Use GPU? (y/n): ").lower() == 'y'
                interactive_ga.current_params.use_gpu = new_value and GPU_AVAILABLE
        else:
            print("Invalid choice. Please try again.")
    
    return interactive_ga

# Run the main demonstration
if __name__ == "__main__":
    # Run comprehensive demonstration
    demo_ga = main()
    
    # Optional: Run interactive parameter adjustment
    print("\n" + "="*60)
    print("INTERACTIVE MODE AVAILABLE")
    print("="*60)
    print("Uncomment the next line to run interactive parameter adjustment:")
    print("# interactive_ga = adjust_parameters_interactively()")
    
    # Uncomment to run interactive mode
    interactive_ga = adjust_parameters_interactively()
```
